# Categorical Data Generation Example

This notebook demonstrates generating categorical data.

```python
from structured_data_generator.core.generators import USER_TEMPLATE, generate_from_template
```

```python
# Generate user data with categorical fields
data = generate_from_template(
    USER_TEMPLATE,
    ["Personal Info", "Preferences", "Device Info"],
    count=100,
    seed=123
)

data.head()
```

```python
# Show value counts for categorical columns
print(data['Gender'].value_counts())
print(data['Device Type'].value_counts())
```
