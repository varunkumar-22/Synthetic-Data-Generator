# Time Series Data Generation Example

This notebook demonstrates generating time series data.

```python
from structured_data_generator.core.generators import IOT_SENSOR_TEMPLATE, generate_from_template
import pandas as pd
```

```python
# Generate IoT sensor data with timestamps
data = generate_from_template(
    IOT_SENSOR_TEMPLATE,
    ["Device Info", "Sensor Readings", "Timestamp Info"],
    count=200,
    seed=456
)

data.head()
```

```python
# Convert timestamp to datetime and plot
data['Timestamp'] = pd.to_datetime(data['Timestamp'])
data.plot(x='Timestamp', y='Temperature C', figsize=(12, 6))
```
